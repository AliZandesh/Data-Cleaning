```python
df['Last_Name'] = df['Last_Name'].str.strip('123._/')
# this line of code is equal to about code block
# The characters '123._/' are passed as an argument to the method, indicating that any occurrence of these characters at the beginning or end of a string should be removed.
df
```

```python
# https://stackoverflow.com/questions/66603854/futurewarning-the-default-value-of-regex-will-change-from-true-to-false-in-a-fu

# df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]', '', regex=True) # first we write this
# df["Phone_Number"] = df["Phone_Number"].apply(lambda x : str(x)) # second
# df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10]) # third
df['Phone_Number'] = df['Phone_Number'].str.replace('nan--', '')
df['Phone_Number'] = df['Phone_Number'].str.replace('Na--', '')

df
```

```python
df[['Street_Address', 'State', 'Zip_Code']] = df['Address'].str.split(',',expand=True)
df

# 123 Shire Lane, Shire
# 93 West Main Street
# 298 Drugs Driveway
# 980 Paper Avenue, Pennsylvania, 18503	
```

```python
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes', 'Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No', 'N')
df
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes', 'Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No', 'N')
df
```

```python
df = df.replace('N/a','')
df = df.replace('NaN','') # Since NaN is an actual value it can not replace with whitespace  
df
```

```python
df = df.fillna('')
df
```

```python
for x in df.index:
    if df.loc[x, 'Do_Not_Contact'] == 'Y':
        df.drop(x, inplace=True)
df        
```

```python
for x in df.index:
    if df.loc[x, 'Phone_Number'] == '':
        df.drop(x, inplace=True)
df    
```
